```python
import pandas as pd
import numpy as np
```

```python
base_path = './main/data/unique/'
```

```python
data_species = pd.read_csv(base_path + 'species_nomix.csv')
data_species
```

```python
# classification 열의 값이 1이면 '개' 2이면 '고양이'로 3이면 '기타'로 변경
data_species.loc[data_species['classification'] == 1, 'classification'] = '개'
data_species.loc[data_species['classification'] == 2, 'classification'] = '고양이'
data_species.loc[data_species['classification'] == 3, 'classification'] = '기타'
```

```python
# '기타'인 값 제거
data_species = data_species[data_species['classification'] != '기타']
data_species
```

```python
# csv 파일로 저장
data_species.to_csv(base_path + 'species_labels.csv', index=False)
```
